# 0. Load and preprocess data

### DataLoad & Pre-process

```python
# Load the Excel file
import pandas as pd
df = pd.read_excel("/content/drive/<your_workspace>/<your_project>/Packaging & Download/Data/250514_Sentiment_raw_4524.xlsx",sheet_name="main")


df1 = df[['Media Title', 'Conversation Stream', 'SenRaw', 'Company', 'Keywords\n(Company)', 'Aspect', 'Keywords\n(Aspect)', 'Sentiment']]


df1 = df1.head(4524)

# sentiment가 N/A인 행 제거
df1 = df1[df1['Sentiment'].notna()]

# 전처리 Company와 Aspect 칼럼에 값이 없는데 Sentiment에는 값이 있는 애들은 버리기
df1 = df1[~((df1['Company'].isna() & df1['Aspect'].isna()) & ~df1['Sentiment'].isna())]
df1
```

```python
# Aspect 라벨링 오류 전처리 (e.g, SAmsung, Apple -> Apple, Plastic -> Plastics, thin- > Thin 등 으로 수정)
df2 = df1.copy()
map_dict = {
    "SAmsung": "Apple",
    "Apple ": "Apple",
    "Plastic": "Plastics",
    "thin": "Thin",
    "Cross-app actions general": "Cross-app actions General",
    "grip": "Grip",
    "seven years of security updates": "Seven years of security updates",
    "Shaomi": "Xiaomi"
}

df2["Aspect"] = (
    df2["Aspect"]
      .str.strip()              # 앞뒤 공백 제거
      .replace(map_dict)        # 일괄 치환
)

df2
```

```python
# Aspect 별 각 Sentiment (Positive, Negative, Neutral) 개수 카운팅
aspect_sentiment_counts = df2.groupby(['Aspect', 'Sentiment']).size().unstack(fill_value=0)
aspect_sentiment_counts
```

#### Rare class filtering

```python
from sklearn.model_selection import train_test_split, StratifiedKFold
from collections import Counter
import pandas as pd

# Aspect는 값이 비었지만, Company에는 값이 있는 경우, Company를 Aspect로 사용
df2.loc[df2['Aspect'].isnull() & df2['Company'].notnull(),'Aspect'] = df2['Company']

# Combo 라벨 만들기
df2['combo'] = df2['Aspect'] + '_' + df2['Sentiment']

# 희귀 조합 필터링 (5개 미만은 제외)
counts = df2['combo'].value_counts()
valid_df   = df2[df2['combo'].isin(counts[counts >= 5].index)].copy()
rare_df    = df2[~df2.index.isin(valid_df.index)].copy()   # 희귀 샘플

# 테스트셋의 비율을 15 %로 설정
train_df, test_df = train_test_split(
    valid_df,
    test_size=0.15,
    stratify=valid_df['combo'],
    random_state=42
)

# 희귀 샘플은 train에만 추가
train_df = pd.concat([train_df, rare_df], ignore_index=True)

print("train size:", len(train_df), "test size:", len(test_df))
print(Counter(train_df['combo']).most_common(5))
print(Counter(test_df['combo']).most_common(5))
```
